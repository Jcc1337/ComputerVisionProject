```python
import pandas as pd
import os

# Load the experiment results
csv_path = 'merged_results_final.csv'

# Check if running from another directory, adjust path if necessary
if not os.path.exists(csv_path):
    script_dir = os.getcwd()
    csv_path = os.path.join(script_dir, 'merged_results_final.csv')

df = pd.read_csv(csv_path)

# Extract specific columns
summary_cols = ['trigger_name', 'ASR', 'PBA_mIoU', 'CBA_mIoU']
performance_summary_df = df[summary_cols].copy()

# Sort by ASR in descending order to see the most effective triggers first
performance_summary_df = performance_summary_df.sort_values(by='ASR', ascending=False).reset_index(drop=True)

# Display the table
display(performance_summary_df)
```

```python
# Calculate min and max for the metrics
range_summary = performance_summary_df[['ASR', 'PBA_mIoU', 'CBA_mIoU']].agg(['min', 'max']).T

# Display the min/max range table
display(range_summary)
```
